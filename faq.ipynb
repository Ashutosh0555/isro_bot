```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
# Scraping from mosdac faq page
```

```python
res= requests.get('https://www.mosdac.gov.in/faq-page')
web =bs(res.content)
datas=web.find_all('div',class_='faq-question-answer')
datas
```

```python
ls=[]
for data in datas:
    ques= data.find('a',class_='active')
    iques=ques.get_text(strip=True)
    answer_div=data.find('div',class_='field-item even')
    ians='no answer found'
    if answer_div:
        ans=answer_div.find('span') or answer_div.find('p')
        if ans:
                ians = ans.get_text(strip=True)
        else:
                
                ians = answer_div.get_text(strip=True)
        
        
    ls.append([iques,ians])
ls
```

```python
import json

faq_data = [{"question": q, "answer": a} for q, a in ls]

with open("mosdac_faq.json", "w", encoding="utf-8") as f:
    json.dump(faq_data, f, indent=4, ensure_ascii=False)
```
